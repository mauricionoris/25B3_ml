# Algoritmos de Classificação - Random Forest

**Random Forest** é um algoritmo de aprendizado supervisionado que pode ser utilizado tanto para problemas de **classificação** quanto de **regressão**, embora seja mais comumente aplicado em tarefas de classificação. Como sabemos, uma floresta é composta por várias árvores, e quanto mais árvores, mais robusta é a floresta. Da mesma forma, o algoritmo Random Forest cria múltiplas árvores de decisão a partir de diferentes amostras dos dados e obtém a predição de cada uma delas, selecionando por fim a melhor solução por meio de **votação**. Trata-se de um método de ensemble (conjunto), que geralmente apresenta desempenho superior a uma única árvore de decisão, pois **reduz o overfitting** (ajuste excessivo) ao realizar uma média dos resultados.

Como observamos, uma das principais desvantagens das árvores de decisão é sua tendência a ajustar-se demais aos dados de treino. O Random Forest é uma maneira de lidar com esse problema. Uma floresta aleatória é, essencialmente, uma coleção de árvores de decisão, onde **cada árvore é ligeiramente diferente das outras**. A ideia por trás do Random Forest é que cada árvore pode realizar uma boa predição individualmente, mas tende a cometer overfitting em partes diferentes dos dados. Ao construirmos várias árvores, todas razoavelmente boas e com diferentes padrões de overfitting, podemos reduzir o overfitting geral ao **calcular a média dos resultados**. Essa redução do overfitting, mantendo o poder preditivo das árvores, pode ser demonstrada matematicamente.

Para aplicar essa estratégia, precisamos construir diversas árvores de decisão. Cada árvore deve ser capaz de prever razoavelmente bem o alvo (target) e, ao mesmo tempo, ser **diferente das demais**. O nome “floresta aleatória” vem justamente da introdução de **aleatoriedade** no processo de construção das árvores, garantindo que elas sejam distintas entre si. Existem duas formas principais de inserir essa aleatoriedade:

1. Selecionando **diferentes pontos de dados** para construir cada árvore.
2. Escolhendo **subconjuntos aleatórios de atributos (features)** em cada divisão (split) da árvore.

Vamos explorar esse processo com mais detalhes a seguir.



## Conjuntos (Ensembles) de Árvores de Decisão

**Ensembles** são métodos que combinam múltiplos modelos de aprendizado de máquina para criar modelos mais poderosos. Existem diversos modelos na literatura de aprendizado de máquina que pertencem a essa categoria, mas dois deles se destacam por sua eficácia em uma ampla variedade de conjuntos de dados para tarefas de **classificação** e **regressão** — ambos utilizando **árvores de decisão** como blocos fundamentais: **Random Forests** e **Gradient Boosted Decision Trees**.

### Random Forests

Como já observamos, uma das principais limitações das árvores de decisão é a tendência ao **overfitting** nos dados de treinamento. O algoritmo Random Forest é uma forma de lidar com esse problema. Uma Random Forest é, essencialmente, uma **coleção de árvores de decisão**, onde cada árvore é ligeiramente diferente das demais.

A ideia por trás do Random Forest é que cada árvore pode realizar uma boa predição individualmente, mas provavelmente irá cometer overfitting em partes diferentes dos dados. Se construirmos muitas árvores, todas funcionando razoavelmente bem, mas com padrões distintos de overfitting, podemos **reduzir o overfitting geral ao fazer a média dos resultados**. Essa redução do overfitting, ao mesmo tempo em que se mantém o poder preditivo das árvores, pode ser demonstrada matematicamente de forma rigorosa.

Para implementar essa estratégia, precisamos construir muitas árvores de decisão. Cada uma deve ser **suficientemente boa** na predição do alvo (target) e, ao mesmo tempo, **diferente das demais**. O nome “Random Forest” vem da **introdução de aleatoriedade** no processo de construção das árvores para garantir sua diversidade. Há duas formas principais de introduzir essa aleatoriedade:

1. Selecionando **amostras aleatórias dos dados** para treinar cada árvore.
2. Escolhendo **atributos (features) aleatórios** em cada ponto de divisão (split) dentro das árvores.



## Algoritmo Random Forest

Podemos entender o funcionamento do algoritmo **Random Forest** com a ajuda dos seguintes passos:

**Passo 1** − Primeiro, seleciona-se **amostras aleatórias** a partir do conjunto de dados original.

**Passo 2** − Em seguida, o algoritmo constrói **uma árvore de decisão para cada amostra**. Depois disso, é feita uma **predição individual** com cada árvore.

**Passo 3** − Neste passo, realiza-se uma **votação** com base nas predições feitas por todas as árvores.

**Passo 4** − Por fim, **a predição mais votada** é escolhida como o **resultado final da classificação**.

2. Em seguida, calculamos a **quantidade de votos** recebida por cada classe (ou valor alvo) predito.

3. Por fim, consideramos como **resultado final da predição** do algoritmo Random Forest a classe (ou valor) que recebeu **maior número de votos**.

### Diferença entre Random Forests e Árvores de Decisão

Vamos comparar **Random Forests** com **árvores de decisão**. Algumas características importantes dessa comparação são as seguintes:

* **Random Forest** é um conjunto de múltiplas árvores de decisão.

* **Árvores de decisão** são computacionalmente mais rápidas em comparação com o Random Forest.

* Árvores de decisão muito profundas podem sofrer de **overfitting**. O Random Forest ajuda a **evitar o overfitting** ao construir árvores a partir de subconjuntos aleatórios dos dados.

* O **Random Forest é mais difícil de interpretar**, enquanto uma **árvore de decisão é facilmente interpretável** e pode ser convertida em regras simples.




## Aplicações do Random Forest

* **Previsão de evasão de clientes**: Empresas podem usar o Random Forest para prever quais clientes têm maior probabilidade de **cancelar um serviço** e, assim, tomar medidas para retê-los. Por exemplo, uma operadora de telefonia pode utilizar esse modelo para identificar clientes que estão usando menos seus aparelhos ou que têm histórico de pagamentos atrasados.

* **Detecção de fraudes**: Random Forest pode ser usado para identificar **transações fraudulentas em tempo real**. Um banco, por exemplo, pode empregar esse modelo para detectar transações feitas a partir de **localizações incomuns** ou com **valores anormalmente altos**.

* **Previsão de preços de ações**: Random Forest pode ser utilizado para **prever preços futuros de ações**. No entanto, é importante observar que esse é um desafio extremamente complexo, e **nenhum modelo será perfeitamente preciso**.

* **Diagnóstico médico**: Pode ajudar médicos no **diagnóstico de doenças**. Por exemplo, um médico pode usar um modelo Random Forest para auxiliar no diagnóstico de **câncer** em um paciente.

* **Reconhecimento de imagens**: O modelo pode ser usado para **reconhecer objetos em imagens**. Por exemplo, um carro autônomo pode utilizar Random Forest para identificar **pedestres e outros veículos** na estrada.

## Classificador Random Forest

Um **Random Forest** é um **metaestimador** que treina diversas **árvores de decisão** em diferentes **subconjuntos do conjunto de dados** e utiliza a **média dos resultados** para melhorar a precisão da predição e controlar o overfitting.

Para construir um modelo de Random Forest, você precisa decidir o número de árvores que deseja criar (parâmetro `n_estimators` da classe `RandomForestClassifier` ou `RandomForestRegressor`).

Suponha que queremos construir 10 árvores. Essas árvores serão **construídas de forma totalmente independente** umas das outras, e o algoritmo fará **escolhas aleatórias diferentes para cada árvore**, garantindo que sejam distintas.

Para construir uma árvore, primeiro é feita uma **amostra bootstrap** dos dados. Ou seja, dos `n_samples` disponíveis, selecionamos aleatoriamente (com reposição) `n_samples` vezes. Isso gera um novo conjunto de dados com o **mesmo tamanho** do original, mas com **algumas amostras repetidas** e **outras ausentes** (cerca de um terço dos dados não aparecerá).

---

### Parâmetros do Random Forest Classifier

* **`n_estimators`**: Número de árvores na floresta.
  Em geral, **mais árvores levam a um melhor desempenho**, mas aumentam o **tempo de processamento**.
  Comece com 100 e aumente conforme necessário.

* **`max_depth`**: Profundidade máxima de cada árvore.
  Árvores mais profundas capturam **padrões mais complexos**, mas também têm **maior risco de overfitting**.
  Experimente valores entre 5 e 15, especialmente para conjuntos de dados menores.

* **`max_features`**: Número de atributos considerados para dividir em cada nó.
  Um valor comum é `'sqrt'` (raiz quadrada do total de atributos).
  Ajuste de acordo com o tamanho do conjunto e a importância das features.

* **`criterion`**: Função usada para medir a qualidade da divisão (`'gini'` ou `'entropy'`).
  O **índice de Gini** costuma ser ligeiramente mais rápido, mas ambos apresentam desempenho semelhante.

* **`min_samples_split`**: Número mínimo de amostras necessário para dividir um nó.
  Valores mais altos **reduzem o overfitting**, mas se forem muito altos, **reduzem a complexidade do modelo**.
  Comece com 2 e ajuste conforme necessário.

* **`min_samples_leaf`**: Número mínimo de amostras em uma folha.
  Semelhante ao parâmetro acima, mas aplicado especificamente aos **nós folha**.
  Comece com 1 e ajuste conforme necessário.

* **`bootstrap`**: Indica se o modelo deve usar **amostragem com reposição** (True ou False).
  A amostragem com reposição (**bootstrapping**) ajuda a melhorar a **variância e generalização** do modelo, mas pode **aumentar um pouco o viés**.



```python
from sklearn.ensemble import RandomForestClassifier
# n_estimators: The number of trees in the forest (default=100)
n_estimators_=50
# criterion: How to split the node in each tree? (Entropy/Gini impurity/Log Loss): default='gini'
classifier = RandomForestClassifier(n_estimators=n_estimators_)
classifier.fit(X_train, y_train)
```

### Importância das Variáveis (Feature Importance)

Por fim, ao observar uma única **árvore de decisão**, é comum que as **variáveis mais importantes** apareçam **mais próximas da raiz** da árvore, enquanto as **menos importantes** aparecem **mais próximas das folhas** (ou podem nem aparecer). Portanto, é possível **estimar a importância de uma variável** calculando a **profundidade média** em que ela aparece nas árvores da floresta.

A biblioteca **Scikit-Learn** calcula isso **automaticamente** para cada variável após o treinamento. Você pode acessar esse resultado através do atributo `feature_importances_`.

Por exemplo, o código abaixo treina um `RandomForestClassifier` com `n_estimators = 50` no conjunto de dados **iris** e imprime a importância de cada atributo. Os resultados mostram que as variáveis mais relevantes são **comprimento** (46%) e **largura da pétala** (35%), enquanto **comprimento** e **largura da sépala** são menos importantes (14% e 3%, respectivamente).

Se quiser, posso complementar com o código Python correspondente. Deseja isso?


```python
print('Features mais importantes for n_estimators =', n_estimators_)
for name, score in zip(iris_dataset["feature_names"], classifier.feature_importances_):
  #print(name, score)
  print('Feature: %s, Score: %.2f%%' % (name, score*100))
```

Por fim, precisamos fazer a predição. Isso pode ser feito com a ajuda do seguinte script:

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
```

```python
result1 = classification_report(y_test, y_pred, target_names=iris_dataset['target_names'])
print("Classification Report:")
print (result1)
```

```python
result2 = accuracy_score(y_test,y_pred)
print("Accuracy: {:.1f}%".format(result2 * 100))
```


### Importância das Variáveis nas Árvores

Em vez de analisar toda a árvore, o que pode ser trabalhoso, existem algumas propriedades úteis que podemos extrair para **resumir o funcionamento da árvore**. O resumo mais comum é a chamada **importância das variáveis** (*feature importance*), que avalia o **quanto cada variável contribui** para as decisões tomadas pela árvore.

A **importância de uma variável** é expressa por um **número entre 0 e 1**, em que:

* **0** significa que a variável **não foi utilizada** em nenhuma divisão da árvore;
* **1** significa que a variável **prediz perfeitamente** o alvo (target).

As importâncias das variáveis **sempre somam 1**, o que permite interpretar os valores como proporções relativas da contribuição de cada atributo para o modelo.



Property **feature_importances_**: The impurity-based feature importances.

The higher, the more important the feature. The importance of a feature is computed as the (normalized) total reduction of the criterion brought by that feature. It is also known as the Gini importance.

```python
model.feature_importances_
```

```python
importance = pd.Series(model.feature_importances_, index = X_train.columns.values)
importance.nlargest(7).plot(kind='barh', figsize = (6,4), title='Feature importances');
```

```python
import numpy as np
from sklearn.metrics import f1_score
from sklearn.model_selection import ParameterGrid
param_grid = {'n_estimators': range(50, 500, 100)}
grid = ParameterGrid(param_grid)
#list
f1score = []
for params in grid:
  model = RandomForestClassifier(n_estimators = params['n_estimators'])
  model.fit(X_train, y_train)
  predict = model.predict(X_test)
  f1 = f1_score(y_test, predict)
  f1score.append(f1)


print("\n Melhor Configuração: \n")
print(grid[np.argmax(f1score)], f1score[np.argmax(f1score)])


```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
forest = RandomForestClassifier(n_estimators=5, random_state=2)
forest.fit(X_train, y_train)
```

```python
y_predict = forest.predict(X_test)
print(classification_report(y_test, y_predict))
acc = accuracy_score(y_test, y_predict)
print("Accuracy: {:.1f}%".format(acc * 100))
```
